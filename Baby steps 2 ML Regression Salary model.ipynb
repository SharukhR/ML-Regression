```python
import pandas as pd #this command imports the pandas library and gives it an alternate name as  pd
```

```python
dataset=pd.read_csv("Salary_Data.csv") # reads csv file and converts it into dataset from file in the same folder and saves it in the variable called dataset
```

```python
independent=dataset[["YearsExperience"]] # seperating input column and saving it under independent variable 
```

```python
dependent=dataset[["Salary"]] # seperating output column and saving it under dependent variable 
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
#importing the train_test_split function from sklearn.model_selection used to split dataset into training and testing parts.we use this to check whether our model is good at what its designed to do.
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
# importing the LinearRegression class from sklearn. It's used to build a linear regression model. we are assigning a variable called 'regressor'. '.fit' is used to train the model using x_train and y_train.
```

```python
weight=regressor.coef_
weight
# '.coef_' represents  the co-efficient or slope .The coefficient represents how much the output (Salary) changes for every one unit increase in input (YearsExperience)
```

```python
bais=regressor.intercept_
bais
#This gives bias of the regression line. bias is the starting point of slope.
```

```python
y_pred=regressor.predict(x_test)
# using the model to make predictions on test data (x_test) to compare the  actual data with predicted output.
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)

#r2_score function is used to measure the accuracy of your regression model.
```

```python
r_score

# r score tells the variation between actual data and predicted output.
```

```python
import pickle 
filename="finalized_model_linear.sav"
# we are saving the model using pickle under the file name "finalized_model_linear.sav"
```

```python
pickle.dump(regressor,open(filename,'wb'))

# saving the model in binary format for to use in future without retraining.
```

```python
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))
result=loaded_model.predict([[15]])
# loading the saved model to make a prediction for 15 years experiance 
```
